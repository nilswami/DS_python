```python
import numpy as np
import pandas as pd
```

#### 1. Load the train.csv as df 

```python
df = pd.read_csv("train.csv")
df
```

#### 2. Drop the irrelevant columns (drop Country column here)

```python
df = df.drop(labels="Country", axis=1)
df
```

#### 3. Rename the below mentioned columns 
* "Life expectancy" :  "Life_Expectancy"
* "Adult Mortality" :  "Adult_Mortality"
* "percentage expenditure" : "percentage_Exp"

```python
dict = {
    "Life expectancy":"Life_Expectancy"
    ,"Adult Mortality":"Adult_Mortality"
    ,"percentage expenditure":"Percentage_Exp"}

df.rename(columns=dict,
          inplace=True)
df
```

#### 4. Drop the rows in the dataset having duplicates in the column "Life_Expectancy"

```python
df.drop_duplicates(subset = "Life_Expectancy", keep = 'first', inplace = True)
df
```

#### 5. Check if there any missing data in the dataset

```python
print(df.isnull().sum())
```

#### 6. Drop the missing data from the dataset

```python
df.dropna(inplace = True)
print(len(df))
```
