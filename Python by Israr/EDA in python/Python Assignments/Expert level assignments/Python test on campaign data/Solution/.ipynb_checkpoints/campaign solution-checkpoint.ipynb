```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_1 = pd.read_excel("CampaignData.xlsx", 'Table 1') # Importing data from table 1
df_2 = pd.read_excel("CampaignData.xlsx", 'Table 2') # Importing data from table 2
```

```python
df = pd.merge(df_1, df_2, on='Campaign') # Merging both the tables in single file
df
```

## 1. Find out the top 5 months in terms of impressions

```python
# Before we try to solve we need seperate the year and thed month from the date function
# Here we are seperating the month and the year from the date column using the lamda function

df['Year']  = df['Date'].apply(lambda x: int(str(x)[:4]))
df['Month'] = df['Date'].apply(lambda x: int(str(x)[5:7]))
df.drop(['Date'],axis=1,inplace=True)
df
```

```python
# Lets try to find the solution using groupby

df.groupby(by = ["Month"])
df
```

```python
# We tried using group by but we need to more creative to find the solutions

# Lets try using for loop to seperate the impressions according to month

topimpressions = df['Month'].unique()
imp = []

for i in topimpressions:
    impressions=len(df[(df['Month'] == i) & (df['Metric'] == 'Impressions')])
    value=np.sum(df[(df['Month'] == i) & (df['Metric'] == 'Impressions')]["Value"])
    imp.append(value)

imp
df1=pd.DataFrame()
df1['Month'] = topimpressions
df1['Impressions'] = imp
df1.head()
```

```python
# As we can see we have got the total impressions according to the month, we just need to find the top 5 now. 

df1.sort_values(by=['Impressions'], ascending=False).head()
```

#### Therefore you can see that the top 5 months in terms of impression are -
* 9
* 3 
* 10
* 8
* 2

## 2.	Which model has maximum impressions?

```python
# Checking the type of models which are preset in the data set
df['Model'].unique()
```

```python
# Here we are using for loop to find the maximum impressions

topmodel = df['Model'].unique()
mod= []

for i in topmodel:
    models=len(df[(df['Model'] == i) & (df['Metric'] == 'Impressions')])
    value= np.sum(df[(df['Model'] == i) & (df['Metric'] == 'Impressions')]["Value"])
    mod.append(value)

mod
df2=pd.DataFrame()
df2['Model'] = topmodel
df2['Impressions'] = mod
df2.head()
```

## 4.	Does the campaign with maximum impressions also have maximum clicks?

```python
# Instead of seperating the impressions and clicks lets try to run the code in a single for loop.

topclicks = df['Campaign'].unique()
camp_3 = [] # I'll create 2 empty list to directly compare the values
clic = []

for i in topclicks:
    impressions=len(df[(df['Campaign'] == i) & (df['Metric'] == 'Impressions')])
    clicks=len(df[(df['Campaign'] == i) & (df['Metric'] == 'Clicks')])
    value=np.sum(df[(df['Campaign'] == i) & (df['Metric'] == 'Impressions')]["Value"])
    value1=np.sum(df[(df['Campaign'] == i) & (df['Metric'] == 'Clicks')]["Value"])
    camp_3.append(value)
    clic.append(value1)

camp_3
clic
df5=pd.DataFrame()
df5['Campaign'] = topclicks
df5['Impressions'] = camp_3
df5['clicks'] = clic
df5.head()
```

```python
# We need to arrange the values in order to find the maximum clicks and the impressions
df5.sort_values(by=['Impressions'], ascending=False).head()
```

```python
df5.sort_values(by=['clicks'], ascending=False).head()
```

## 5.	Find out the top 3 campaigns in terms of CPM (CPM stands for cost per mile which means cost per 1000 impressions. CPM = total media cost/total impressions * 1000)

```python
campaign = df['Campaign'].unique()

cpm=[]
# value=0
for i in campaign:
    impressions=len(df[(df['Campaign'] == i) & (df['Metric'] == 'Impressions')])
    value=np.sum(df[(df['Campaign'] == i) & (df['Metric'] == 'Media Cost')]["Value"])
#     values+=value
    cpm_i = value*1000/impressions
    cpm.append(cpm_i)
    
    
cpm
df6=pd.DataFrame()
df6['Campaign'] = campaign
df6['CPM'] = cpm
df6.head()
```

```python
df6.sort_values(by=['CPM'], ascending=False).head()
```

#### Therefore the top 3 campaigns in terms of CPM 

               Campaign 	        CPM
    7 	Sentra campaign 11 	  1.548958e+06
    8 	Sentra campaign 14 	  1.393852e+06
    12 	Sentra campaign 7 	  8.083346e+05

## 8.	When would you use a Pie chart, a Bar chart and a Line Chart? Use the given dataset and provide an example for each type of chart.

* When you want to find the percentage of different categories within each column for categorical data we use Pie chart. 
* But when you want to compare each category against each other within a column for a categorical dataset we use a bar chart. 
* Line chart is effective when the column values need to be analyzed based on time (month, quarters, year)
