```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import calendar
```

#### 1. Create a data frame and name it df1

```python
df1 = pd.DataFrame()
```

#### 2. Create 5 columns and name the columns as (a,b,c,d,e)

```python
df1 = pd.DataFrame(columns=['a','b','c','d','e'])
df1.head()
```

#### 3. In column a, create a sequence of 1-16 with a step size of 2

```python
df1["a"] = pd.Series(np.arange(1,16,2))
df1
```

#### 4. In column b, sequence of starting 8 letters in capital

```python
test_list = []

# for filling alphabets
alpha = 'A'
for i in range(0, 8):
    test_list.append(alpha)
    alpha = chr(ord(alpha) + 1) 

df1["b"]= test_list

df1
```

#### 5. In column c, create a sequence of starting 8 months

```python
df1["c"] = [calendar.month_name[i] for i in range(8, 0, -1)]
df1
```

#### 6. In column d create a sequence as M1, M2, M3,....M8

```python
col = []
counter = 1

for i in range(len(df1)):
    value = i+1
    if value == 9:
        break
    else:
        key = 'M' +str(counter)
        counter +=1
        col.append(key)

df1["d"] = col
df1
```

#### 7. In column e, create a sequence of starting 8 letters in small 

```python
test_list = []

# for filling alphabets
alpha = 'a'
for i in range(0, 8):
    test_list.append(alpha)
    alpha = chr(ord(alpha) + 1) 

df1["e"]= test_list

df1
```

#### 9. Create 5 columns and name the columns as (a,b,c,d,e)

#### 10. In column a, create a sequence of 1-36 with a step size of 3

#### 11. In column b, create a sequence of starting 9 to 20 letters in capital

```python
test_list = []

# for filling alphabets
alpha = 'I'
for i in range(0, 12):
    test_list.append(alpha)
    alpha = chr(ord(alpha) + 1) 

df2["b"]= test_list

df2
```

#### 12. In column c, create a sequence of 12 months 

```python
df2["c"] = [calendar.month_name[i] for i in range(1, 13)]
df2
```

#### 13. In column d, create a sequence as C1,C2,C3,....C12

```python
col = []
counter = 1

for i in range(len(df2)):
    value = i+1
    if value == 13:
        break
    else:
        key = 'C' +str(counter)
        counter +=1
        col.append(key)

df2["d"] = col
df2
```

```python
test_list = []

# for filling alphabets
alpha = 'i'
for i in range(9, 21):
    test_list.append(alpha)
    alpha = chr(ord(alpha) + 1) 

df2["e"]= test_list

df2
```
