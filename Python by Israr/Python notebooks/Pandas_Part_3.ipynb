# Itertating over a dataframe
1. itertuple
2. iteritems
3. iterrows

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
df = pd.DataFrame(space_org)
df
```

```python
# itertuple
# 
for row in df.itertuples():
    print(row)
```

```python
# itertuple
# 
for row in df.itertuples():
    print(row.Organizations_name)
```

```python
# itertuple
# 
for row in df.itertuples():
    if row.Formation_year == 1993:
        print(row.Headquarters)
```

```python
# itertuple
for row in df.itertuples():
    if row.Organizations_name == 'ISRO':
        print(row.Country_name)
```

```python
# itertuple
for i in df.iterrows():
    if i[1].Organizations_name == 'ISRO':
        print(i[1].Country_name)
#    print(i[1].Organizations_name)
```

```python
# itteritems
for row in df.iteritems():
    print(row,end = '\n\n')
```

```python
df
```

```python
for i in df.iteritems():
    if i[0] == 'Country_name':
        print(i[1])
```

```python
# WORKING
for i in df.iteritems():
    if i[0] == 'Country_name':
        print(i[1].loc[1])
        
```

```python
# itertuple
for i in df.iterrows():
    if i[1].Organizations_name == 'ISRO':
        print(i[1].Country_name)
#    print(i[1].Organizations_name)
```
