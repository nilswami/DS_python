```python
# Points to keep in mind

1. To read a file a file you need to open a file

syntax for file handling - 
 - f = open('file_name.ext', 'mode')
 - # work
 - f.close()

closing the is important otherwise what ever changes you have made will not take place


2. data read from the file is always in a string format
3. mode of operation - text files

- read mode - 'r'
- write mode - 'w'
- append mode - 'a'
- read and write mode - 'r+'
- write and read mode - 'w+'
- append and read mode - 'a+'

4. mode of operation - binary files

- read mode - 'rb'
- write mode - 'wb'
- append mode - 'ab'
- read and write mode - 'r+b'
- write and read mode - 'w+b'
- append and read mode - 'a+b'


5. methods on file objects

- read() -- reading the content from the file
- write() -- writing a content in the file
- tell() -- tells the cursor position 
- seek() -- moves the cursor to the specific index location
```

# Modes of operation 

```python
## 1. Read mode - 'r'
- If the file exists then it will open the file in read mode and places the cursor at the begining of the file
- if the file does not exists then it will give FileNotFoundError
- You are allowed only to read from the file but not allowed to write into the file
```

```python
ls
```

```python
# if the file does not exists then it will give FileNotFoundError
f = open('read_mode.txt' , 'r')
```

```python
ls
```

```python
# file is present
f = open('read_mode.txt' , 'r')
print(f.tell())
print(f.read())
print(f.tell())
print(f.read())
print(f.seek(4))
print(f.read())
```

2. Write mode - 'w'

* If the file exists then it will open the file in write mode and deletes the content of the file
* if the file does not exists then it will create a file with given name and extension
* You are allowed only to write into the file but not allowed to read from the file

```python
# If the file does not exist

f = open ('write_mode.py', 'w')
```

```python
# If the file does exist

f = open ('write_mode.py', 'w')
print(f.tell())
print(f.write('hey guys'))
print(f.tell())
print(f.seek(2))
print(f.read())
f.close()
```

```python
#
f = open ('write_mode.py', 'w')
print(f.tell())
print(f.write('hey guys'))
print(f.tell())
print(f.seek(0))
print(f.write('hello'))
print(f.tell())
f.close()
```

## 3. Append mode - 'a'
- If the file exists then it will open the file in write mode and places the cursor at the end of the file
- if the file does not exists then it will create a file with given name and extension
- You are allowed only to write into the file but not allowed to read from the file
- append will always write the the content at the end of the file so seek() will not work here

```python
# if the file is not present 
f = open('append_mode.txt', 'a')
```

```python
# if the file is not present 
f = open('append_mode.csv', 'a')  #\n , \t
print(f.tell())
print(f.write('\n'))
print(f.tell())
print(f.write('hardik'))
print(f.seek(2))
print(f.tell())
print(f.write('paris'))
f.close()
```

# 6. Append and read mode - 'a+'

- if the file does not exists then it will create a file with given name and extension
- You are allowed to write into the file and also allowed to read from the file
