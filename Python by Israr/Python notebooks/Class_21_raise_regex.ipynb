```python
class error_amt(Exception):
    def __init__(self,msg):
        self.msg = msg
```

```python
class error_pan(Exception):
    def __init__(self,msg):
        self.msg = msg
```

```python
def submit_pan(pan):
    if len(pan) > 7:
        print('We will verify it soon')
    else:
        raise error_pan ('Wrong pan number')

def amount_deposit(amt):
    if amt > 50000:
        raise error_amt('Over limit')
    else: 
        print('Amount deposited')
    
def amount_deposti_after_check(amt):
    if amt > 100000:
        raise error_amt('Over limit')
    else: 
        print('Amount deposited')
```

```python
# if the amount was greater than 50000
# Invalid pan number
amt = int(input('enter the amount'))
try:
    amount_deposit(amt)
except error_amt as e:
    print(e)
    try:
        pan = input('Please enter the pan card ')
        submit_pan(pan)
        amount_deposti_after_check(amt)
    except error_pan as e:
            print(e)
```

```python
'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
'''
```
