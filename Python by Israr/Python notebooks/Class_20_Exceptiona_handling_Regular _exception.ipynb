```python
class MyException(Exception):
    def __init__(self,msg):
        self.msg = msg

def submit_pan(pan):
    if len(pan)==7:
        print('We will verify it soon')
    else:
        raise MyException('Wrong pan number')

def amount_deposit(amt):
    if amt > 50000:
        raise MyException('Over limit')
    else: 
        print('Amount deposited')
```

```python
amt = int(input('enter the amount'))
try:
    amount_deposit(amt)
except MyException as e:
    print(e)
    try:
        pan = input('Please enter the pan card ')
        submit_pan(pan)
    except MyException as e:
            print(e)
```

# Regular expession

```python
where we can search or match specific text
```

```python
'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
'''
```

```python
import re 
```

```python
text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ\s
321-555-4321
1234567890
Ha HaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
mohammed.israr11.com
321-555-4321
123.555.1234
123*555*-1234
123.555.1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
Mr_hello
'''
```

```python
# what you want to match
pattern = re.compile('great')
# From which text you have to match 
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# dot(.)

# what you want to match
pattern = re.compile('.')
# From which text you have to match 
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \d (0-9)
# To extract the degits 
# what you want to match
pattern = re.compile('\d')
# From which text you have to match 
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \D      - Not a Digit (0-9)
pattern = re.compile('\D')
# From which text you have to match 
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \w - (a-z , A-Z , 0 - 9, _)

pattern = re.compile('\w')
# From which text you have to match 
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \W (not a word character)

pattern = re.compile('\W')
# From which text you have to match 
matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \s whitespace ( space, tab \t , \n)

a = re.compile('\s')

# From which text you have to match 

matches = a.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \S not a white space

a = re.compile('\S')

# From which text you have to match 

matches = a.finditer(text_to_search)

for match in matches:
    print(match)
```
